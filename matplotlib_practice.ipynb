```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis,y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# plt.plot(x_axis, y_axis)
plt.plot(x_axis, y_axis, label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0, 50)
# plt.grid()
plt.title("PyBer Fare by Month")
plt.legend()
# plt.show()
plt.savefig("./Image.png")
```

```python
# Size to 20 pixels wide by 10 pixels high
plt.figure(figsize=(20, 10))
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Size to 20 pixels wide by 10 pixels high
plt.figure(figsize=(10, 5))
# Create the plot.
plt.plot(x_axis, y_axis, marker="o", color="red", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Set rabge January - June
plt.xlim("Jan", "June")
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Matplotlib Object-Oriented Functions
# Change the size of the figure in pixels.
fig, ax = plt.subplots(figsize=(10, 5))
# Add a label that will be added to the legend.
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Sets the min and max range of the y-axis.
ax.set_ylim(0, 50)
# Sets the min and max range of the x-axis.
# ax.set_xlim(0, 10)
# Add a label to the x-axis.
ax.set_xlabel('Date')
# Add a label to the y-axis.
ax.set_ylabel('Fare($)')
# Add a title.
ax.set_title("PyBer Fare by Month")
# Add a legend.
ax.legend()
# Add a grid to the chart.
ax.grid()
# Saves the figure with the given extension.
plt.savefig("./Image2.png")
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# Invert Axis
plt.gca().invert_yaxis()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Invert months
plt.gca().invert_yaxis()
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create bubble Chart
fig, ax = plt.subplots()
# Color sky blue, 5 times each data point, 20% transparency, black edge color, linewidth 2 points, legend Boston
ax.scatter(y_axis, x_axis, color = "skyblue", s = [i * 5 for i in y_axis], alpha = 0.80, edgecolor="black", linewidth=2, label = "Boston")
# Show legend
ax.legend()
# Set Title
ax.set_title("PyBer Fare by Month")
# Add a label to the x-axis.
ax.set_xlabel('Fare($)')
# Add a label to the y-axis.
ax.set_ylabel('Date')
# Sets the min and max range of the y-axis.
ax.set_xlim(0, 50)
# Invert axis
ax.invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, explode=explode_values, colors = colors, labels=x_axis, autopct="%.1f%%")
plt.show()
```
